# A Statistical Analysis on Inference in Old Age

## Introduction

In the following experiment, the average of correct responses to questions requiring inferential answers, between young adults and older adults, will be compared and evaluated to determine whether the ability to make correct inferences differ across age. As a bases for this experiment, it has been stated within the literature that "information processing is constrained by the amount of mental capacity at the given moment", coming from the limited capacity model of cognitive functioning (Zacks et al, 1987). With this model, the goal of this experiment is to faciliate inference making by having participants read passages that require them to use information obtained through the discourse of each passage. 

The design of this experiment is to present a passage that has a intended inference but contains a competing inference that would conflict with the original. This is depicted by having the first paragraph display context supporting the competing inference, while having the second paragraph provide supporting context to the intended inference. In having the two inferences presented in this way, when the participant comes across the second paragraph, they will encounter new information that conflicts with the context presented in the first paragraph. The occurrence of this confliction would result in extra demand on one's mental capacity, having "one to refer back to their memory of the preceding paragraph to find a basis for the correct inference, seeking a coherent representation of the given passage" (Zacks et al, 1987). This in part will be used to determine whether older adults experience a deficit in their ability to make correct inferences, due to their strained mental capacity. As a result, the following analysis presents to find whether inference generation is consistent between both age groups.


**Scientific Question:** Does there exist an age difference in inference generation.

**Main Goal**: To determine whether older adults retain their ability to make correct inferences as their younger counterparts.

>**Hypothesis Test**

>$H_{O}$: The mean of correct inferences for older adults is equal to the mean of correct inferences for younger adults.

>$H_{A}$: The mean of correct inferences for older adults not equal to the mean of correct inferences for younger adults.

As a means of determining whether there exist a significant difference between age groups, this experiement will set the null hypothesis to be the assumption that the means of both age groups are equal, while the alternative hypothesis will be set to be the assumption that the means of both age groups are significantly different. 

***

## Methods

###  Participants

For the sake of the experiment, 100 participants were sampled from the general public. 50 of the participants consisted of young adults ranging from 18 to 35 years of age, while the other 50 particpants were of older adults ranging from 60 to 70 years of age. Members of each group were selected by their WAIS-R Vocabulary Test scores, using the test as a pre-screen. A certain thresold was set to ensure that all participants were above a certain level of competency. 

>* 50 Young Adults (Mean Age = 20 Years, Range 18-35)
>* 50 Old Adults (Mean Age = 70 Years, Range 60-70)

***

### Materials and Design

In terms of the materials and design of this experiment, there will be 18 passages for each participant to read, each depicting a scene or an event which is described within two paragraphs. 12 of the passages will contain a competing inference, having the intended inference be the unexpected inference, while the remaining 6 will only contain context in support of the intended inference, serving as filler passages. Six out of the 12 passages were taken from Zacks et al 1987, and the other six consisted of new passages written in the same design. The 6 filler passages were taken from Zacks et al 1987 as well.

Following each passage will be a set of 3 questions where 1 out of the 3 will be a question that requires an inferential answer whereas the other two will require factual answers, acting as filler questions. 

***

There will be 18 passages for each the subjects to read, as well as a set of 3 questions for each passage to answer. 
* One question requires an inferential answer, while the other two require factual answers (Control Questions). For this experiment, we're only looking at the correctness of the inferential questions. 

Each of the passages depicts a scene or an event which provides contexual information to a particular inference.
* 12 out of the 18 passages contain context that would lead to an unexpected inference, a conflicting interpretation**(Competing Inference)**, against the intended**(Target)** inference.


* 6 out of the 18 passages contain context that would lead to an expected inference that reinforced the intended**(Target)** inference. These are used as filler passages.

***

### Procedure
Subjects were tested individually. Each were given 18 passages like the ones stated above, as well as a set of 3 questions for each of passages to answer via an online survey, conducted within a lab. 

Participants' responses were encoded as 1 if their inferential answer matched the intended inference and 0 otherwise.


## Experiement Results

In simulating the data for this experiment, a samplier function was created to sample from a bernoulli distribution. The function takes in a probability as its input in which it uses to obtain 12 samples with 0 or 1 as their values, the probability being the chance of success, or in other words, getting a 1. Another function was used in conjunction with the samplier, where it obtains 50 experiments, returning a list of values where each represent the sum of all correct inferences made within an experiment. The probabilities passed into these functions are the probabilities that an age group gets an inference question correct. In the case of this simulation, the probability for young adults was set to be 90% where as for the older adults, it was set to 88%. The difference between these probabilities is to simulate a possible difference in correct inference making in terms of the two age groups. It will be through a t-test in which the significance of this difference will be deemed significant or not.

***

```r
# Simulating Test Scores
bern.samplier <- function(prob){rbinom(1:12, 1, prob)}

samplier <- function(prob){
    test_scores = list()
    for (i in 1:50){
        test_scores[i] <- list(bern.samplier(prob))
    }
    group_scores <- sapply(test_scores, sum)
    return(group_scores)
}

# Probabilities
young.prob = .90
old.prob = .88

young_group_results <- samplier(young.prob)
old_group_results <- samplier(old.prob)

# Results for both groups
young_group_results
young_group_average = mean(young_group_results)
cat("Young Adult Average:", young_group_average, "\n\n")

old_group_results
old_group_average = mean(old_group_results)
cat("Older Adult Average:", mean(old_group_results))

# Plotting the distribution of correct inferences between the two groups
options(repr.plot.width=4, repr.plot.height=4)
hist(young_group_results, xlab = "Number of Correct Inferences", main = "Young Adults Inference Distribution", ylim = c(0,25))
hist(old_group_results, xlab = "Number of Correct Inferences", main = "Old Adults Inference Distribution",  ylim = c(0,25))
```

## Statistical Analysis

In performing the statistical analysis, a two-sample t-test was conducted to determine whether the difference between means will prove to be significant or not. As seen below, the squared error for older adults was calculated, along side the variance, standard deviation and squared error. With these components, a t-statistic was calculated where it was compared to the critical t-value. Based on the critical t-value being -1.66 and the t-statistic -1.22, it has come to show that there isn't a statistically significant between the two age groups since the t-statistic doesn't GO BEYOND -1.66, leading one to fail to reject the null hypothesis. 

```r
squared.error = function(mean)
{
    function(x)
    {
        (mean-x)^2
    }
}

# Standard Error
compute.squared.error = squared.error(young_group_average)
sq.error = compute.squared.error(old_group_average)
cat("Squared-Error:", sq.error)

# Estimated Variance
sampled.square.errors = sapply(old_group_results, compute.squared.error)
sum.sampled.square.errors = sum(sampled.square.errors)
num.samples = length(old_group_results)
estimated.variance = sum.sampled.square.errors / (num.samples - 1)
cat("\nEstimated Variance: ", estimated.variance)

# Standard Deviation
estimated.sd = sqrt(estimated.variance)
cat("\nEstimated Deviation: ", estimated.sd)

# Standard-Error
standard_error = estimated.sd / sqrt(num.samples)
cat("\nStandard-Error: ", standard_error)

# t-statistic
t.statistic = (old_group_average - young_group_average) / standard_error
cat("\nt-statistic: ", t.statistic)

# Critial t-value
cat("\nCritical t-value (degrees of freedom = 97, alpha = 0.05): ", -1.66)
```

***
By performing another test, its confirmed that the difference proves to be insignificant, having a p-value of 0.3896, or a 39% chance of seeing the difference that we have seen given that the null hypothesis is true.

```r
# Two sample T-test
t.test(old_group_results, young_group_results)
```

## Misleading Results

In having the first experiment reveal that there isn't a significant difference between the means of each age group, it has been decided that the same experiment will be conducted multiple times. It is when one of the experiments show a significant difference between the two means where the findings will be set to be published. The drive behind this is publication bias, where once results that are desired are obtained, one neglects to dispute the possibility that the significant results were obtained by chance. 

As seen below, the same experiement is set to run 10 times, or until a significant p-value is obtained. Based on the results, it appears that the experiment ran three times before doing so, acquiring a p-value of .04, showing to be significant. With the critical t-value being -1.66, having a t-value of -1.7232 confirms the fact that based on these results, the null hypothesis can be rejected. Misleadingly, we can state that there is a difference in terms of correct inference making between the two age groups, when in actuality, there isn't a significant difference at all.

```r
young.prob = .90
old.prob = .88

# Repeats the experiment 10 times, or at least until a p-value < .05 is achieved
for (iter in 1:10){
    young_group_results <- samplier(young.prob)
    old_group_results <- samplier(old.prob)

    p_value <- t.test(old_group_results, young_group_results, alternative = "less")[3]
    
    if (p_value <= .05){
        cat("Number Experiments conducted until reaching a significant result:", iter)
        print(t.test(old_group_results, young_group_results, alternative = "less"))
        # Plotting the distribution of correct inferences between the two groups
        options(repr.plot.width=4, repr.plot.height=4)
        hist(young_group_results, xlab = "Number of Correct Inferences", main = "Young Adults Inference Distribution", ylim = c(0,25))
        hist(old_group_results, xlab = "Number of Correct Inferences", main = "Old Adults Inference Distribution",  ylim = c(0,25))
        break
    }
}
```
